##### 2. urllib.error 异常处理模块

```python
import urllib.request
import urllib.error
import socket
try:               # try: except: 异常处理
    res = urllib.request.urlopen('http://httpbin.org/get',timeout = 0.1)
    print (res.read().decode('utf-8'))
except urllib.error.URLError as e:                 #只能用别名，直接写会出错
    print (str(e.reason) == 'timed out')
    print (isinstance(e.reason,socket.timeout))
    print ('失败')
```

##### 5. Requset对象

Request类:

```python
from urllib import request as e
req = e.Request('http://httpbin.org/get')
res = e.urlopen(req)
print (res.read().decode('utf-8'))
print (res.getheader('server'))
```

```python
import urllib.parse
from urllib import request as r
url = 'http://httpbin.org/post'
data = bytes(urllib.parse.urlencode({'time':'16:59'}),encoding='utf-8')
headers={
    'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'
}
requ = r.Request(url = url , data = data, headers = headers , method = 'post')
res = r.urlopen(requ)
print (res.getheaders())
```

采用 add_header方法添加头 添加多个多次调用就可以

```python
import urllib.parse
from urllib import request as r
url = 'https://docs.python.org/3/library/http.cookiejar.html'
#data = bytes(urllib.parse.urlencode({'time':'16:59'}),encoding='utf-8')
requ = r.Request(url = url)
requ.add_header('User-Agent','Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36')
res = r.urlopen(requ)
print (res.read().decode('utf-8'))
```

#### cookie

```python
import urllib.request

res = urllib.request.urlopen('http://www.baidu.com')
print (res.getheader('Set-Cookie'))
```

```python
import urllib.request
import http.cookiejar
cookie = http.cookiejar.CookieJar()          #前面必须携带http
hander = urllib.request.HTTPCookieProcessor(cookie)   #返回hander句柄，创建cookie处理器
opener = urllib.request.build_opener(hander)          #创建opener
res = opener.open('http://www.baidu.com')
for e in cookie:                       #返回的为http.cookiejar里面的CookieJar类
    print (e.name + " = " + e.value)         #类中含有的name、value属性
```

```python
from urllib import request
from http import cookiejar
cookie = cookiejar.CookieJar()
hander = request.HTTPCookieProcessor(cookie)
opener = request.build_opener(hander)
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'
}
requ = request.Request('https://www.jianshu.com/p/28c3a9cd8e3c',headers = headers)
res = opener.open(requ)
print ('cookies:')
for item in cookie:
    print (item.name + '=' + item.value)
```

```python
from urllib import request
from http import cookiejar
file_name = 'cookie.txt'
cookie = cookiejar.MozillaCookieJar(file_name) #直接使用MozillaCookieJar
#cookie = cookiejar.CookieJar()
handle = request.HTTPCookieProcessor(cookie)
opener = request.build_opener(handle)
res = opener.open('http://www.baidu.com')
for item in cookie:
    print (item.name + ' = ' + item.value)
cookie.save(ignore_discard = True , ignore_expires = True)
```

#### 练习 post 、 cookie post

保存

```python
from urllib import request
from http import cookiejar
file_name = 'MyCookie.txt'
cookie = cookiejar.MozillaCookieJar(file_name)
handle = request.HTTPCookieProcessor(cookie)
opener = request.build_opener(handle)
res = opener.open('http://httpbin.org/cookies/set/tom/19')
cookie.save(ignore_discard = True , ignore_expires = True);
```

读取、发送

```python
from urllib import request
from http import cookiejar
file_name = 'MyCookie.txt'
cookie = cookiejar.MozillaCookieJar(file_name)
cookie.load(file_name , ignore_discard = True , ignore_expires = True)
handle = request.HTTPCookieProcessor(cookie)
opener = request.build_opener(handle)
resu = request.Request(url = 'http://httpbin.org/cookies')
res = opener.open(resu);
print (res.read().decode('utf-8'))
```
